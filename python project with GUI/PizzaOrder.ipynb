```python
class Pizza:
    def get_description(self):
        return self.desc
    def get_price(self):
        return self.desc
    class Klasik:
        name = "Klasik Pizza"
        desc = "Ben Klasik Pizzayım."
        price = 70;
    class Margarita:
        name = "Margarita Pizza"
        desc = "Ben Margarita Pizzayım."
        price = 50
    class TurkPizza:
        name = "Türk Pizza"
        desc = "Ben Türk Pizzayım."
        price = 75
    class Sade:
        name = "Sade Pizza"
        desc = "Ben Sade Pizzayım."
        price = 40
```

```python
class Soslar:
    def get_description(self):
        return self.desc
    def get_price(self):
        return self.desc
    class Zeytin:
        name = "Zeytin"
        desc = "Ben Zeytinim."
        price = 5;
    class Mantar:
        name = "Mantar"
        desc = "Ben Mantarım."
        price = 6
    class KeciP:
        name = "Keçi Peyniri"
        desc = "Ben Keçi Peyniriyim."
        price = 7
    class Et:
        name = "Et"
        desc = "Ben Etim."
        price = 8
    class Sogan:
        name = "Soğan"
        desc = "Ben Soğanım."
        price = 9
    class Misir:
        name = "Mısır"
        desc = "Ben Mısırım."
        price = 10
```
